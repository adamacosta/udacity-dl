```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
```

```python
def random_images(n=4):
    return [np.random.ranf(784).reshape((28, 28)) for _ in range(n)]

def show_images(imgs):
    """Create a random image to visualize the difference from MNIST digits."""
    n = len(imgs)
    f, axarr = plt.subplots(1, n, sharey=True, figsize=(16, 4))
    for i in range(n):
        axarr[i].imshow(imgs[i])
    plt.show()
```

```python
# Any randomly selected point in our 784-dimension looks like noise
show_images(random_images())
```

```python
# Compare this to MNIST images
mnist = input_data.read_data_sets("MNIST_data/")
imgs, _ = mnist.train.next_batch(4)
show_images(imgs.reshape((4, 28, 28)))
```
